## Project Title:
Hallmark-Driven Gene Expression Signatures of Angiogenesis and Growth Suppression Evasion in Prostate and Ovarian Cancer


## Project Goal:
This project aims to investigate whether cancer hallmarks, specifically sustained angiogenesis and evading growth suppression, are predictive of patient survival and disease characteristics in aggressive, metastasis-prone cancers. We focus on prostate adenocarcinoma (PRAD) and ovarian serous carcinoma (OV) to explore how hallmark-associated gene expression relates to survival outcomes, tumor stage, and treatment response, and whether hallmark expression patterns differ across cancer types.


## Disease Background:
Cancer Hallmark Focus

Sustained Angiogenesis
Evading Growth Suppression

Overview of Sustained Angiogenesis
Sustained angiogenesis enables tumors to grow beyond diffusion limits by inducing the formation of new blood vessels. Tumors hijack normal vascular signaling pathways to increase oxygen and nutrient supply, facilitating rapid growth and metastasis.
Key genes involved:

VEGFA, FGF2, ANGPT2, HIF1A, TNF

Overview of Evading Growth Suppression
Cancer cells evade growth suppression by disrupting tumor suppressor pathways, allowing uncontrolled proliferation. This often occurs through mutations in genes responsible for cell cycle regulation, DNA damage response, and apoptosis.
Key genes involved:

TP53, RB1, PTEN, ARID1A, CDKN2A (p16), RUNX3

Cancer Types of Interest
Prostate Adenocarcinoma (PRAD)

Common cancer with highly variable outcomes
Advanced or metastatic PRAD spreads rapidly to bone and lymph nodes
Alterations in PTEN, TP53, and RB1 are associated with aggressive disease and resistance to therapy

Ovarian Serous Carcinoma (OV)

One of the deadliest gynecologic cancers due to late detection
Characterized by rapid spread throughout the peritoneal cavity
Nearly universal TP53 disruption and strong reliance on angiogenic signaling pathways

Prevalence & Mortality

PRAD: High incidence, lower mortality overall, but poor outcomes in metastatic disease
OV: Lower incidence, significantly higher mortality due to late-stage diagnosis

Risk Factors & Societal Determinants

Genetic alterations (e.g., TP53, BRCA-associated pathways)
Age, access to screening and care, comorbidities
Disparities in early detection and treatment access influence outcomes

Standard of Care Treatments

PRAD: Androgen deprivation therapy, chemotherapy, radiation
OV: Surgical debulking, platinum-based chemotherapy, anti-angiogenic therapies 

Biological Mechanisms 
Both cancers exploit angiogenic signaling to sustain growth and acquire mutations that disable growth-suppressive checkpoints, enabling rapid proliferation and metastasis.

The gene expression data and metadata were first loaded and subsetted to include only PRAD and OV cancer samples. PCA was used to explore patterns in the data after standardizing gene expression values. The PCA plot colored by cancer type showed clear separation between PRAD and OV samples, suggesting that these hallmark genes capture meaningful biological differences between the two cancers.

Similarly, when colored by growth suppression score, the PCA plot showed shifts in score values across the PCA space, suggesting that growth suppression activity also contributes to differences between samples.





K-Means clustering was applied using two clusters to identify natural groupings in the data. The resulting clusters closely matched the separation seen between PRAD and OV samples in the PCA plot. Overall, these results suggest that the selected hallmark genes contain useful information and support moving forward with a supervised learning model to classify cancer types.

Conclusions

This study used hallmark driven gene expression profiles to explore biological differences between prostate adenocarcinoma (PRAD) and ovarian serous carcinoma (OV). By focusing on genes associated with sustained angiogenesis and evading growth suppression, the analysis emphasized biologically meaningful features rather than genome wide expression alone. Principal Component Analysis revealed clear separation between PRAD and OV samples in low dimensional space, indicating that these hallmark pathways capture dominant sources of variation between the two cancer types. Unsupervised clustering using K means further supported this distinction, with cluster assignments closely matching cancer type despite no labels being used during training. Together, these results suggest that angiogenesis and growth suppression pathways play distinct and measurable roles in structuring gene expression patterns across cancers with different metastatic behaviors. Overall, this work demonstrates that biologically informed dimensionality reduction can reveal meaningful cancer specific structure and supports the use of supervised learning methods in future analyses.

Ethical Implications

The use of gene expression based machine learning approaches in cancer research raises several ethical considerations. Public genomic datasets such as TCGA may reflect underlying demographic and healthcare access biases, which can limit the generalizability of findings to more diverse patient populations. There is also a risk of overinterpreting exploratory or correlational results, particularly when unsupervised methods are used, which could lead to misleading conclusions if findings are not adequately validated. In addition, although TCGA data are de identified, the use of genomic information more broadly raises ongoing concerns related to patient privacy, data security, and informed consent. Ethical application of these methods therefore requires transparency regarding model limitations, validation across independent cohorts, and careful framing of results as exploratory rather than clinically actionable.

Several limitations should be considered when interpreting the results of this analysis.
