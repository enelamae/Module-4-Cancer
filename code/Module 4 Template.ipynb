For this analysis, we used gene expression and clinical metadata from the GSE62944 TCGA training dataset,focusing on Prostate Adenocarcinoma (PRAD) and Ovarian Serous Cystadenocarcinoma (OV)- one known to typically metastasize slowly, whereas the other is known to metastasize aggressively (Regional Cancer Care Associates, 2025; The University of Kansas Cancer Center, 2020). Our analysis centers on two specific hallmarks of cancer: angiogenesis (represented by VEGFA, HIF1A, ANGPT1, ANGPT2, and FLT1) and the evasion of growth suppressors (captured via TP53, RB1, CDKN2A, PTEN, and SMAD4).

In the gene expression data, each row corresponds to a gene (e.g., VEGFA, TP53), and each column represents an individual tumor sample. Expression values are reported as log₂-transformed TPM, which provides a normalized measure of mRNA abundance across each samples. The accompanying metadata links each sample to clinical characteristics such as cancer type, age, and sex. All data were generated through the TCGA project, where tumor samples were sequenced using RNA-seq. Raw read counts were normalized to TPM and then log₂-transformed to reduce variance and allow for more consistent comparison of gene expression patterns across the two cancer types.
