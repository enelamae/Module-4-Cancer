## Project Title:
Hallmark-Driven Gene Expression Signatures of Angiogenesis and Growth Suppression Evasion in Prostate and Ovarian Cancer


## Project Goal:
This project aims to investigate whether cancer hallmarks, specifically sustained angiogenesis and evading growth suppression, are predictive of patient survival and disease characteristics in aggressive, metastasis-prone cancers. We focus on prostate adenocarcinoma (PRAD) and ovarian serous carcinoma (OV) to explore how hallmark-associated gene expression relates to survival outcomes, tumor stage, and treatment response, and whether hallmark expression patterns differ across cancer types.


## Disease Background:
Cancer Hallmark Focus

Sustained Angiogenesis
Evading Growth Suppression

Overview of Sustained Angiogenesis
Sustained angiogenesis enables tumors to grow beyond diffusion limits by inducing the formation of new blood vessels. Tumors hijack normal vascular signaling pathways to increase oxygen and nutrient supply, facilitating rapid growth and metastasis.
Key genes involved:

VEGFA, FGF2, ANGPT2, HIF1A, TNF

Overview of Evading Growth Suppression
Cancer cells evade growth suppression by disrupting tumor suppressor pathways, allowing uncontrolled proliferation. This often occurs through mutations in genes responsible for cell cycle regulation, DNA damage response, and apoptosis.
Key genes involved:

TP53, RB1, PTEN, ARID1A, CDKN2A (p16), RUNX3

Cancer Types of Interest
Prostate Adenocarcinoma (PRAD)

Common cancer with highly variable outcomes
Advanced or metastatic PRAD spreads rapidly to bone and lymph nodes
Alterations in PTEN, TP53, and RB1 are associated with aggressive disease and resistance to therapy

Ovarian Serous Carcinoma (OV)

One of the deadliest gynecologic cancers due to late detection
Characterized by rapid spread throughout the peritoneal cavity
Nearly universal TP53 disruption and strong reliance on angiogenic signaling pathways

Prevalence & Mortality

PRAD: High incidence, lower mortality overall, but poor outcomes in metastatic disease
OV: Lower incidence, significantly higher mortality due to late-stage diagnosis

Risk Factors & Societal Determinants

Genetic alterations (e.g., TP53, BRCA-associated pathways)
Age, access to screening and care, comorbidities
Disparities in early detection and treatment access influence outcomes

Standard of Care Treatments

PRAD: Androgen deprivation therapy, chemotherapy, radiation
OV: Surgical debulking, platinum-based chemotherapy, anti-angiogenic therapies 

Biological Mechanisms 
Both cancers exploit angiogenic signaling to sustain growth and acquire mutations that disable growth-suppressive checkpoints, enabling rapid proliferation and metastasis.
